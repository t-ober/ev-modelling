```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
```

```python
mu1, sigma1 = 1, 0.1
data1 = np.random.normal(mu1, sigma1, 10000)
```

```python
plt.figure()
count, bins, ignored = plt.hist(data1, 40, density=True)
```

```python
mu2, sigma2 = 2, 0.3
data2 = np.random.normal(mu2, sigma2, 10000)
```

```python
plt.figure()
count, bins, ignored = plt.hist(data2, 40, density=True)
```

```python
data = np.concatenate((data1, data2))
```

```python
plt.figure()
count, bins, ignored = plt.hist(data, 80, density=True)
```

```python
from sklearn.mixture import GaussianMixture 
```

```python
gm = GaussianMixture(n_components=2, n_init=10).fit(data.reshape(-1,1))
```

```python
gm.means_
```

```python
gm.covariances_
```

```python
gm.weights_
```

```python
x = np.linspace(0, 3, 10000)
```

```python
logprob = gm.score_samples(x.reshape(-1,1))
```

```python
responsibilities = gm.predict_proba(x.reshape(-1,1))
```

```python
pdf_ges = np.exp(logprob)
```

```python
pdf_individual = responsibilities * pdf_ges[:, np.newaxis]
```

```python
pdf_individual
```

```python
plt.figure()
count, bins, ignored = plt.hist(data, 80, density=True, alpha=0.8)
plt.plot(x, pdf_ges, '-k')
plt.plot(x, pdf_individual, '--k')
```
